```python
import numpy as np
from funktioner import LinearRegression
```

### Läs in och rensa datasetet

```python
data = np.genfromtxt(
    "housing.csv",
    delimiter=",",
    skip_header=1,
    usecols=range(9)
)

data = data[~np.isnan(data).any(axis=1)]
```

### Dela upp datasetet

```python
X = data[:, :-1]
Y = data[:, -1]
```

### Skapa och träna modellen

```python
model = LinearRegression(X, Y)
model.fit()
```

### Visa antal datapunkter och features, och beräkna och visa varians, standardavvikelse och RMSE

```python
print("Antal datapunkter (n):", model.n)
print("Antal features (d):", model.d)
print("Varians:", model.variance())
print("Standardavvikelse:", model.std())
print("RMSE:", model.rmse())
```
